```python
# fetch all news about WMT
walmart_headlines = newsapi.get_everything(
    q="walmart",
    language="en",
    page_size=100,
    sort_by="relevancy"
)

# print total articles
print(f"Total articles about Walmart: {walmart_headlines['totalResults']}")

# show sample article
walmart_headlines["articles"][0]
```

```python
# get sentiment of single article

print(walmart_headlines['articles'][1]['description'])

# view the score of the article's description
analyzer.polarity_scores(walmart_headlines['articles'][1]['description'])
```

```python
# create dataframe with sentiment of all news articles 

walmart_sentiments = []

for article in walmart_headlines["articles"]:
    try:
        text = article["content"]
        date = article["publishedAt"][:10]
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment["compound"]
        pos = sentiment["pos"]
        neu = sentiment["neu"]
        neg = sentiment["neg"]
        
        walmart_sentiments.append({
            "text": text,
            "date": date,
            "compound": compound,
            "positive": pos,
            "negative": neg,
            "neutral": neu
            
        })
        
    except AttributeError:
        pass
    
# Create DataFrame
walmart_df = pd.DataFrame(walmart_sentiments)

# Reorder DataFrame columns
cols = ["date", "text", "compound", "positive", "negative", "neutral"]
walmart_df = walmart_df[cols]

walmart_df.head()
```
